###  Column Descriptions

| Column | Description |
|--------|------------|
| Index | Unique identifier for each video game record. |
| Name | The name of the video game.|
| Platform | The platform on which the game is available, such as PC, PS4, Xbox, etc. |
| Year_of_Release | The year in which the game was released. |
| Genre | The genre of the game, such as Action, Sports, etc.|
| Publisher|  The company that published the game.  |
| NA_Sales| The sales of the game in North America, in millions. |
| EU_Sales| The sales of the game in Europe, in millions.|
| JP_Sales| The sales of the game in Japan, in millions. |
| Other_Sales| The sales of the game in other regions, in millions.|
| Global_Sales| The total global sales of the game, in millions. |
| Critic_Score| The average score given to the game by critics.|
| Critic_Count| The number of critics who reviewed the game.|
| User_Score| The number of critics who reviewed the game.|
| User_Count | The average score given to the game by users.|
| Developer| The company that developed the game.|
| Rating| The ESRB rating of the game, such as E for Everyone, T for Teen, or M for Mature.|

## Repairing Columns:

#### Detecting and fixing problems in dataset columns such as inconsistent formats, wrong data types, or invalid entries.

```python
repair_columns=['year_of_release', 'publisher', 'na_sales', 'eu_sales', 'jp_sales','other_sales','global_sales','user_score']
```

```python
""" 
    - Convert all object dtype to string dtype: optimized internally, making operations faster and more memory efficient.
    - year_of_release: contain 'Adventure'(invalid entry) will replace it with NaN (becuase it is wrong value that is belong to year) 
                     and convert the data type from float into int (ex: from [2006.] to [2006]), then fill it.
                   
    - publisher: need to remove the puncatiations: , - // . to be better.
    - na_sales, eu_sales, jp_sales, other_sales and global_sales: need to convert the scientific notation (like 4.136e+01) into standard decimal formatting (like 41.36) to be clear and more readable.
    - user_score: need to replace 'tbd' to NaN and convert the data type from object into float, then fill it. 
      
"""
```

Convert all object dtype to string dtype

```python
# Identify object columns
obj_cols = df.select_dtypes(include='object').columns
print(obj_cols)
```

```python
# Convert them to pandas string dtype
df[obj_cols] = df[obj_cols].astype("string")

# Verify conversion
print(df.dtypes)
```

```python
# Repair 'year_of_release' column

# Step1: Convert 'year_of_release' to numeric, invalid entries (like: Adventure) → NaN
df['year_of_release'] = pd.to_numeric(df['year_of_release'], errors='coerce') 

# Step2: convert to Int to remove float decimals (e.g., 2009.0 → 2009).
df['year_of_release'] = df['year_of_release'].astype('Int64')  

# Check unique values
df['year_of_release'].unique()
```

```python
# Repair the 'publisher' column: need to remove the puncatiations: '-' , '.' , '//' , ',' to be better.
df['publisher'] = (
    df['publisher']
    .str.strip()
    .str.lower()
    .str.replace('-', ' ')
    .str.replace('.', ' ')
    .str.replace('//', ' ')
    .str.replace(',', ' ')
)
```

```python
# Run it after repair to see the changes:
print('publisher') 
print(df['publisher'].nunique())
print(df['publisher'].unique())
```

```python
# Repair cols_to_format
cols_to_format=['na_sales','eu_sales','jp_sales','other_sales','global_sales']

# Define function for decimal formatting.
def decimal_format(x, decimals=2):
    if pd.isna(x):   # to keep NaN as it is.
        return x
    else:
        return format(float(x), f'.{decimals}f')
```

```python
# Apply formatting function.
for col in cols_to_format:
    df[col]=df[col].apply(decimal_format)
```

```python
for col in cols_to_format:  # to check the output
    print(col) 
    print(df[col].nunique())
    print(df[col].unique(), '\n','*'*100)
```

Calculate the outliers to fill the numericall columns

```python
def outlier_percentage(series):
    series = series.dropna()
    
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = series[(series < lower_bound) | (series > upper_bound)]
    
    return round(len(outliers) / len(series) * 100, 2)

fill=['user_score','year_of_release']

outlier_report = {
    col: outlier_percentage(df[col])
    for col in fill
}


print(f'outlier_report: {outlier_report}')
```

```python
for col in fill:
    fig_box= px.box(df,x=col, title=f"box plot of {col}")
    fig_box.show()
```

#### Additional Features to Enhance Understanding and better explanation
###### Sales Category: Define a function to categorize sales and apply it in the existing global_sales to create a new feature named sales_category.
###### Game Age Category: Calulate game age and define a function to categorize game age is it new, recent, old or classic.



```python
# Sales Category

df['global_sales'] = pd.to_numeric(df['global_sales'], errors='coerce')   # to ensure global_sales is numeric

def categorize_sales(value):
    
    if 0 <= value < 25:
        return 'low'
    elif 25 <= value < 50:
        return 'medium'
    elif 50 <= value < 75:
        return 'high'
    elif 75 <= value <= 100:
        return 'best_seller'
    else:
        return 'out_of_range'  # optional for values > 100

# Apply the function to create a new column
df['sales_category'] = df['global_sales'].apply(categorize_sales)

print(df[['name','global_sales','sales_category']].head())
```

```python
# Game Age Category

current_year= datetime.datetime.now().year
df['game_age']=current_year - df['year_of_release']      # Calulate Gmae age

print(df[['year_of_release','game_age']].head())
```

```python
def categorize_age(value):
    
    if value < 5:
        return 'new'
    elif 5 <= value < 15:
        return 'recent'
    else:
        return 'old'
    
       

# Apply the function to create a new column
df['game_age_category'] = df['game_age'].apply(categorize_age)

print(df[['game_age_category','game_age']].head())
```
